```python
import ipywidgets as widgets
from ipywidgets import FileUpload
import matplotlib.pyplot as plt

```

```python
uploader = FileUpload(
accept = '*.csv', 
multiple = False)
display(uploader)
```

```python
import io
import pandas as pd
input_file = list(uploader.value.values())[0]
content = input_file['content']
content = io.StringIO(content.decode('utf-8'))
df = pd.read_csv(content)
```

```python
df
```

```python
df.columns
```

```python
df.plot.line('Year', 'Mean', figsize=(20,12))
```

```python
import seaborn as sns
```

```python
df0= df.drop('Source', axis=1)
```

```python
df1= df0.sort_index(axis=0, ascending=True)
```

```python
df1 = df1.set_index('Year')
```

```python
plt.figure(figsize=(13,7))
ax=sns.heatmap(df1.T, cmap='coolwarm')## we can transpose df1 to make it look better
ax.set_title('Heatmap of the global annual temperature change')
```
